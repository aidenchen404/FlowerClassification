```python
import matplotlib.pyplot as plt
import numpy as nps
```

```python
np.random.seed(18977878)  # seed the random number generator.
data = {'a': np.arange(50), #various ways of getting "random" numbers
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 10

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 5), layout='constrained')
ax.plot(x, x, label='striahgt')  # Plot some data on the Axes.
ax.plot(x, x**2, label='cruve')  # Plot more data on the Axes...
ax.plot(x, x**4, label='cruvier')  # ... and some more.
ax.set_xlabel('People be like "What a wonderful graph!"')  # Add an x-label to the Axes.
ax.set_ylabel('Meanwhile, "the graph:" ')  # Add a y-label to the Axes.
ax.set_title("Wow, is this a (Matplotlib Figure) ?!")  # Add a title to the Axes.
ax.legend()  # Add a legend.
```

```python
from matplotlib.colors import LogNorm

X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 + X**9 + Y**4) * np.exp(-X**2 - Y**2)
fig, axs = plt.subplots(layout='constrained')
pc = axs.imshow(Z**2 * 100, cmap='plasma', norm=LogNorm(vmin=0.01, vmax=100))
fig.colorbar(pc, extend='both')
axs.set_title('imshow() with LogNorm()')
```
